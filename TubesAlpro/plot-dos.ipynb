```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
plt.style.use('style/sci.mplstyle')
import numpy as np
figsize = (12, 5)
dpi = 600 

# Open and read the file gr.dos
x , y = np.loadtxt('pendoutb.dat', unpack=True)
a, b = np.loadtxt('pendout2b.dat', unpack=True)
    
# Create figure object
plt.figure(figsize=figsize, dpi=dpi)
# Plot the DOS, in which the Fermi energy shifts to zero
plt.plot(x,y, color ='red')
plt.plot(a,b, color ='blue')
# Add the x and y-axis labels
plt.xlabel('Time (t)')
plt.ylabel('theta/omega')
# Set the axis limits
#plt.xlim(-20, 20)
#plt.ylim(0, 4)
# Save the figure to the pdf file
plt.savefig('plot-dos.pdf')
```
